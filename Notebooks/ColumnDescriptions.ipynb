0. Catalog : Tells what catalog the data is from (all H indicating Hipparcos)

1. HIP : Unique identifier for an individual star (likely not useful information)

2. Proxy : Flag indicating an object from the Hipparcos or Tycho catalog is nearby. H and T should be replaced with a 'Present' flag, and nulls should be replaced with an 'Absent' flag

3. RAhms : Right Ascension measured in hours, minutes, seconds. The units of time are equivalent to degrees of movement, so this should be split and then summed together into a measure of 'RA in seconds'

4. DEdms : Declination measured in degrees. DEdeg is likely more usable, as it is already a usable numeric column which has the same information.

5. Vmag : Magnitude measured in Johnson V. No data cleaning needed

6. VarFlag : Flags variability type, only present where variability is known.

7. r_Vmag : Source of Vmag, could be onehotencoded to see if it's useful

8. RAdeg : Right ascension in degrees, a different way to measure RAhms; likely only use one at a time.

9. DEdeg : Declination in degrees, a different way to measure DEdms, already numeric.

10. AstroRef : Reference flag for astrometric params. 80-97% nulls, likely not useful; but maybe try a missing indicator with onehotencoding.

11. Plx : Trigonometric parallax; Parallax measures the apparent motion against the backdrop of more distant objects. No data cleaning needed.

12. pmRA : Proper motion measured in Right ascension. Proper motion is movement of a star through the sky. No cleaning needed.

13. pmDE : Proper motion in Declination. No cleaning needed, but don't use with pmRA, only one is needed since they measure the same thing.

14. e_RAdeg : Standard error of RAdeg.

15. e_DEdeg : Standard error of DEdeg.

16. e_Plx : Standard error of Plx.

17. e_pmRA : Standard error of pmRA.

18. e_pmDE : Standard error or pmDE.

19. DE:RA : Correlation coefficient of DE and RA.

20. Plx:RA : Correlation coefficient of Plx and RA.

21. Plx:DE : Correlation coefficient of Plx and DE.

22. pmRA:RA : Correlation coefficient of pmRA and RA

23. pmRA:DE : Correlation coefficient of pmRA and DE

24. pmRA:Plx : Correlation coefficient of pmRA and Plx

25. pmDE:RA : Correlation coefficient of pmDE and RA

26. pmDE:DE : Correlation coefficient of pmDE and DE

27. pmDE:Plx : Correlation coefficient of pmDE and Plx

28. pmDE:pmRA : Correlation coefficient of pmDE and pmRA

29. F1 : Percentage of data rejected. No cleaning needed.

30. F2 : Goodness of fit param. Indicates how normally distributed data is, 0 indicating very normal data, +/- 3 indicating less normal data, indicating a poor fit.

31. --- : Repeated HIP number, drop column.

32. BTmag : Another measure of magnitude.

33. e_BTmag : Standard error of BTmag

34. VTmag : Another measure of magnitude.

35. e_VTmag : Standard error of VTmag

36. m_BTmag : Flag indicating which part of a multi star system was measured for the BTmag. More useful as a binary with nulls be filled as 'Non multi system' and present values filled as 'Multi system' to be one hot encoded.

37. B-V : Johnson bv color. A measure of wavelengths of light (color), purely numeric, just needs imputing.

38. e_B-V : Standard error of B-V

39. r_B-V : Source of bv color index, likely not useful, but could be onehotencoded with nulls imputed as most frequent.

40. V-I : Color index in cousins system

41. e_V-I : Standard error in V-I

42. r_V-I : Source of color index, some letters can be combined into fewer groups, which can then be onehotencoded

43. CombMag : Indicates for non single stars which component was measured for bt and vt. Can be made binary

44. Hpmag : Median magnitude.

45. e_Hpmag : Standard error of Hpmag

46. Hpscat : Scatter of H_P observations

47. o_Hpmag : Number of observations for hpmag median, standard error, and scatter

48. m_Hpmag : Multi star system flag, as other reference flags.

49. Hpmax : Magnitude at max luminosity

50. HPmin : Magnitude at min luminosity

51. Period : Variability period, not usable for predicting variability.

52. HvarType : Target, indicates variability type.

53. moreVar : Additional data about variability, not usable for predictions.

54. morePhoto : Light curve annex. Not usable for predicting variability as it is always null when a star is variable.

55. CCDM : Catalog of Components of Double and Multiple Stars identifier. May be useful to make a nullflag.

56. n_CCDM : Indicates whether the star was classified as part of a multi star system prior to the Hipparcos satelite. 

57. Nsys : Number of entries with the same CCDM; indicates how many stars are in the system.

58. Ncomp : Number of components entry was resolved into; binary flag could be useful

59. MultFlag : Double and Multiple Systems annex. May be most useful as a binary flag of 'multiple star system or not'

60. Source : Source of MultFlag, not enough data to be useful.

61. Qual : Indicates certainty with which a star is classified as part of a multiple star system or not. Potentially very useful, but many nulls

62. m_HIP : Component designation for multi system columns theta, rho, and e_rho.

63. theta : Angle between components specified in m_HIP (Multi system)

64. rho : Angular separation between components specified in m_HIP (Multi system)

65. e_rho : Standard error in rho

66. dHp : Magnitude difference of components specified in m_HIP

67. e_dHp : standard error in dHp

68. Survey : Indicates a survey star; a list of particularly bright stars. Contains 2 values; S and Null, turn to binary if used

69. Chart : Flag indicating an identification chart. Many nulls, likely not useful; but could do a null indicator/onehotencode

70. Notes : Indicates existence of notes on the star. Mostly null

71. HD : Id number for HD/HDE/HDEC

72. BD : BD Identifier.

73. CoD : CoD Identifier

74. CPD : CPD Identifier

75. (V-I)red : Color index used for photometric processing

76. SpType : Spectral type aquired from ground based compilations, largely taken from Hipparcos Catalog. Some updates have been made, largely to variable stars

77. r_SpType : Source of SpType from a variety of catalogs. Has a flag for updated after publication of HIC.
